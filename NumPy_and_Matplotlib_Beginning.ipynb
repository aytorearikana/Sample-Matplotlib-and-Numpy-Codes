Matplotlib

Helps us to draw graphics with numerical values

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
N = 300
x= np.linspace(0 , 2*np.pi*10 , N)
y = np.cos(x)
plt.figure(figsize =(8,5))
plt.plot(x,y, 'r')
plt.show()
```

```python
N = 250
x= np.linspace(0 , 2*np.pi*10 , N)
y = np.cos(x)*np.exp(x/5)
plt.figure(figsize =(8,4))
plt.plot(x,y, 'r')
plt.show()
```

```python
N = 250
x= np.linspace(0 , 2*np.pi*5 , N)
y = np.cos(x)*np.exp(- x/5)
plt.figure(figsize = (8,4))
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('np.cos(x)*np.exp(x/5)')
plt.show()
```

```python
N = 500
x= np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
y=np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
#WE DEFINE R BECAUSE WE WANT TO TURN AROUND OUR GRAPHICS AROUND RADIAN
Z = np.cos(R)*np.exp(-R/5)

plt.figure(figsize = (8,8))
plt.contour(X,Y,Z)
```

```python
N = 500
x= np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
y=np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
#WE DEFINE R BECAUSE WE WANT TO TURN AROUND OUR GRAPHICS AROUND RADIAN
Z = np.cos(R)*np.exp(-R/5)

plt.figure(figsize = (8,8))
plt.contourf(X,Y,Z)
```

3D

```python
from mpl_toolkits import mplot3d
#to draw the graphic in 3d
```

```python
N = 500
x= np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
y=np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
#WE DEFINE R BECAUSE WE WANT TO TURN AROUND OUR GRAPHICS AROUND RADIAN
Z = np.cos(R)*np.exp(-R/5)

plt.figure(figsize = (8,8))
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z,cmap='jet')
```

```python
N = 50
x= np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
y=np.linspace(-2*np.pi*5, 2*np.pi*5 , N)
X,Y = np.meshgrid(x,y)
Z = np.cos(X)*np.sin(Y)

plt.figure(figsize = (8,8))
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z,cmap='jet')
```
